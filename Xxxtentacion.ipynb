```python
import requests, json, urllib.parse

base_url = 'https://itunes.apple.com/search'

artist = 'Xxxtentacion'

params = {
    'term': artist,
    'limit': 200,
    'media': 'music',
    'entity': 'song',
    'attribute': 'artistTerm'
}

response = requests.get(f"{base_url}?{urllib.parse.urlencode(params)}")

data = json.loads(response.text)
data = data['results']
# Все доступные ключи объекта
print(list(data[0].keys()))

import dateutil.parser

# Преобразование строки с датой в объект даты
# **x - скопировать все поля из объекта x
data = map(lambda x: {**x, 'releaseDate': dateutil.parser.isoparse(x['releaseDate'])}, data)
data = list(data)

# Убираем лишние поля
data = map(lambda x: {
    'artistId': x['artistId'],
    'collectionId': x['collectionId'],
    'trackId': x['trackId'],
    'artistName': x['artistName'],
    'collectionName': x['collectionCensoredName'],
    'trackName': x['trackCensoredName'],
    'collectionPrice': x['collectionPrice'],
    'trackPrice': x['trackPrice'],
    'releaseDate': x['releaseDate'],
    'collectionExplicitness': x['collectionExplicitness'],
    'trackExplicitness': x['trackExplicitness'],
    'discCount': x['discCount'],
    'discNumber': x['discNumber'],
    'trackCount': x['trackCount'],
    'trackNumber': x['trackNumber'],
    'trackTimeMillis': x['trackTimeMillis'],
    'country': x['country'],
    'currency': x['currency'],
    'primaryGenreName': x['primaryGenreName'],
    'isStreamable': x['isStreamable']
}, data)
data = list(data)
print(list(data[0].keys()))

# Фильтрация с целью оставить искомого артиста
data = filter(lambda x: artist.lower() in x['artistName'].lower(), data)
data = list(data)
def convert_millis(millis: int) -> str:
    total_seconds = millis // 1000
    seconds = total_seconds % 60
    minutes = (total_seconds // 60) % 60
    hours = (total_seconds // (60 * 60))
    return f"{hours}:{minutes}:{seconds}"
```

```python
from statistics import mean, median
name = data[3]['artistName']
track_num = len(set(map(lambda x: x['trackId'], data)))
genre = max(list(map(lambda x: x['primaryGenreName'],data)), key=list(map(lambda x: x['primaryGenreName'],data)).count)
album_count = len(set(map(lambda x: x['collectionId'], data)))
total_length = sum(list(map(lambda x: x['trackTimeMillis'], data)))
mean_length = round(mean(list(map(lambda x: x['trackTimeMillis'], data))))
median_length = round(median(list(map(lambda x: x['trackTimeMillis'], data))))
longest_song_name = max(data, key=lambda x: x['trackTimeMillis'])['trackName']
longest_song_time = max(data, key=lambda x: x['trackTimeMillis'])['trackTimeMillis']
shortest_song_name = min(data, key=lambda x: x['trackTimeMillis'])['trackName']
shortest_song_time = min(data, key=lambda x: x['trackTimeMillis'])['trackTimeMillis']

total_cost = sum(list(map(lambda x: x['trackPrice'], data)))
mean_cost = round(mean(list(map(lambda x: x['trackPrice'], data))))
median_cost = round(median(list(map(lambda x: x['trackPrice'], data))))
expensive_song_name = max(data, key=lambda x: x['trackPrice'])['trackName']
expensive_song_price = max(data, key=lambda x: x['trackPrice'])['trackPrice']
cheapest_song_name = min(data, key=lambda x: x['trackPrice'])['trackName']
cheapest_song_price = min(data, key=lambda x: x['trackPrice'])['trackPrice']

print(name)
print(track_num)
print(genre)
print(album_count)
print(convert_millis(total_length))
print(convert_millis(mean_length))
print(convert_millis(median_length))
print(longest_song_name)
print(convert_millis(longest_song_time))
print(shortest_song_name)
print(convert_millis(shortest_song_time))
print(round(total_cost))
print(mean_cost)
print(median_cost)
print(expensive_song_name)
print(expensive_song_price)
print(cheapest_song_name)
print(cheapest_song_price)
```
